### Recommendations:

The most important thing that the Red Cross should focus more on is adding more doctors, nurses, and beds. There were about 2 doctors on average and 3 nurses on average at both health facilities and points of interest. Some facilities have no doctors even though there are more than 10 beds and some clinic facilities have neither doctors nor nurses.  During a pandemic this would not be enough to help patients or respond to the surge of patients that need care, as hospitals can be overrun and cause more spread to the disease.  The allocation of doctors and nurses should be based on the hospital bed count, and a proper ratio of bed:doctor:nurse should be considered.

Another recommendation is that health facilities should have a greater rate of staying open 24/7. Currently only about 25% of all facilities open 24/7 and less than 25% of all hospitals open 24/7. It can be targeted to have all hospitals open 24/7. During a pandemic, clinics and hospitals should be ready to treat patients at any time of the day.

Furthermore, many hospitals and clinics still have no electricity or generators as power sources.  I recommend to upgrade them to solar power and they should be run by the government.  

The survey revealed that the Red Cross should focus on improving quality and accessibility of drinking water and toilet facilities. Without clean drinking water and toilet facilities, the spread of disease can rapidly increase. This is because more people would share water and cause pollution, which will expose people to one another. Without toilet facilities, sanitation will be at an all time low, which will spread more diseases.

With all the above, Red Cross can build a better and cleaner environment for people and avoid the spread of diseases.  Red Cross can establish a quick response to allow sick people get proper treatment.
